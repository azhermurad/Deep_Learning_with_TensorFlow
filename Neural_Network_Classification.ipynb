```python
import numpy as np

def plot_decision_boundary(model, X, y):

  # Define the axis boundaries of the plot and create a meshgrid
  x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
  y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
  xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                       np.linspace(y_min, y_max, 100))

  # Create X values (we're going to predict on all of these)
  x_in = np.c_[xx.ravel(), yy.ravel()] # stack 2D arrays together: https://numpy.org/devdocs/reference/generated/numpy.c_.html

  # Make predictions using the trained model
  y_pred = model.predict(x_in)
  y_pred = np.round(tf.squeeze(y_pred).numpy())

  print(y_pred)
  # Check for multi-class
  # if model.output_shape[-1] > 1: # checks the final dimension of the model's output shape, if this is > (greater than) 1, it's multi-class
  #   print("doing multiclass classification...")
  #   # We have to reshape our predictions to get them ready for plotting
  #   y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
  # else:
  #   print("doing binary classifcation...")
  #   y_pred = np.round(np.max(y_pred, axis=1)).reshape(xx.shape)
  y_pred = y_pred.reshape(xx.shape)

  # Plot decision boundary
  plt.contourf(xx, yy,y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
  plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(), xx.max())
  plt.ylim(yy.min(), yy.max())
```

```python

# plot_decision_boundary
plot_decision_boundary(model,X,y)

```

# LearningRateScheduler
At the begin of every epoch this function get the update learning rate from scheduler A function that takes an epoch index (integer, indexed from 0) and current learning rate (float) as inputs and returns a new learning rate as output (float).

```python
tf.random.set_seed(42)
# create the artificail netural Network for the classification problem

model_seven = Sequential(
    [
        Input(shape=(2,)),
        Dense(4,activation='relu'),
        Dense(4,activation='relu'),
        Dense(1,activation='sigmoid')  # sigmoid activation function for binary classfication
    ]
)

model_seven.compile(
    loss = tf.keras.losses.BinaryCrossentropy(),
    optimizer = optimizers.Adam(),
    metrics = ['accuracy']


)

# model fit
def lr_schedule(epoch,lr):
   return 0.001 * (10 ** (epoch / 20))  # Example: decaying every 10 epochs

# Callback to print the learning rate at each epoch
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lr_schedule)

# training_finished = False
class LearningRatePrinter(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        lr = self.model.optimizer.learning_rate
        # print(f"Epoch {epoch + 1}: Learning rate = {lr.numpy()}")


history_seven=model_seven.fit(X_train,y_train,epochs=100,callbacks=[lr_scheduler])
```

## Classification Evaluation Methods

Different Evaluat ion Methods are listed below:
+ Accuracy ( Right prediciton by the model! misleading in dataset is imbalance)
+ Precision (focus on the positive predicted value how many of them will be actual positive (less Fasle positive (model predict positive but actual the value is false))) we use precsion in such case in which we have to reduce the false positive
+ Recall (recall is used when the false negative is important in which the model predict negative but the actual value is positive) recall is used in medical diagonoics  
+ F1-Score (we use f1-score when the false positive and false negative same importance)
+ confusion Metrix (confusion metric compare the Actual values and Predicted values by the model)


### Three ways to calculate the classification methods like precision, recall, accuracy and f1-score etc.


```python
# accuracy which is used for calculate the total true prediction by the model
y_pred = model_eight.predict(X_test)
y_pred = (y_pred > 0.5).astype(int)

accuracy = tf.keras.metrics.Accuracy()
accuracy(y_test,y_pred).numpy()


```
